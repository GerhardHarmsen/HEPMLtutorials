```python
path="/Users/rousseau/Downloads"

outputdatapath="/Users/rousseau/Google\ Drive/GD_openData/dataWW_ATLAS_openData13TeV_filtered/"

!ls -lrt {outputdatapath}
outputdatapath=os.path.abspath(outputdatapath).replace("\ ", " ")  # try to normalise the path (annoyance with the space)
print ("Will write data file to : ",outputdatapath)
```

```python


# cross section to come from https://github.com/atlas-outreach-data-tools/atlas-outreach-cpp-framework-13tev/blob/master/Plotting/inputfiles/Files_HWW.txt
# the one in the input file can be wrong
labels=["qq","ggH","VBFH"]
detaillevel="d1"
cross_sections={"qq":12.466,"ggH":0.58,"VBFH":0.04} # in fb

outputfilename=os.path.join(outputdatapath,"dataWW_"+detaillevel+".csv.gz")

# total integrated luminosity ATLAS 2012 : 139fb^-1
luminosity=10 #fb^-1 about 2015 data 



import os

all_events={}
for label in labels:        
    filepath=os.path.join(path,"dataWW_"+label+"_"+detaillevel+".csv.gz")
    all_events[label] = pd.read_csv(filepath) 


#qq_events = pd.read_csv("/Users/rousseau/Google Drive/GD_opendata/dataWW_ATLAS_openData13TeV_filtered/dataWW_qq_d0_1000k.csv.gz",sep="," ) 
#ggh_events = pd.read_csv(datapath+"dataWW_ggH_d0_100k.csv.gz",sep=",") 
#vbfh_events = pd.read_hdf(datapath+"dataWW_VBFH.hdf","VBFH",WRITEABLE=False ) #Vector Boson Fusion http://opendata.cern.ch/record/3826
all_events["VBFH"].columns
display(all_events["VBFH"].head())
display(all_events["VBFH"].describe())


```

```python
# compute normalisation factors
for label in labels:
    sumw=all_events[label]["mcWeight"].sum()
    totalw=all_events[label]["SumWeights"][0]
    weight_correction=1/sumw
    print("For file ",label," n events ", all_events[label].shape[0]," sum of weights:",sumw, " expected:",totalw," weight correction",weight_correction)
    scaling=luminosity*cross_sections[label]*weight_correction
    print (" now multiply all weights by ",scaling)
    all_events[label]["mcWeight"]*=scaling
```

```python
# add new label for each dataset
all_events["qq"]["label"] = 0 # qq background
all_events["ggH"]["label"] = 1 # ggH signal
all_events["VBFH"]["label"] = 1 # VBFH signal
```

```python
# merge the dataset
fulldata = pd.concat([all_events["qq"],all_events["ggH"],all_events["VBFH"]]) # merge all dataset
# randomise
fulldata=fulldata.sample(frac=1).reset_index(drop=True) 
display(fulldata.describe())
display(fulldata.columns)
```

```python
#dop some columns
fulldata.drop(columns=["XSection","SumWeights"],inplace=True)

#reorder some columns
varlist=fulldata.columns.tolist()
oldvarlist=varlist.copy()

#some stuff to the end
for aname in ["channelNumber"]:
    if not aname in varlist:
        continue
    varlist.remove(aname)
    varlist.append(aname)
    
#some stuff at the beginning
varlist.remove("label")
varlist.insert(1,"label")    

# check we have only reordered
if sorted(oldvarlist)!=sorted(varlist):
    print ("before:",sorted(oldvarlist) )
    print ("after :",sorted(varlist) )
    raise Exception("something wrong when reordering the columns!")

# do the reordering
fulldata = fulldata.reindex(columns=varlist)
display(fulldata.head())
```

```python
#write final dataset 
fulldata.to_csv(outputfilename,float_format='%.5g',index = False)# do csv file
```
